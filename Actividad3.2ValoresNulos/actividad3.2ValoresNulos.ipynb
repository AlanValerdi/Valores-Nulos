```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .csv sin indice
# carga de valores utilizando una variable llamada data
data = pd.read_csv("madrid_spain.csv")
```

```python
# utilizo una variable llamada columns para seleccionar las columnas que quiero en mi dataframe para posteriormente guardarlas en una variable llamada data2
columns = ['id','last_scraped', 'source', 'name', 'description',
                    'neighborhood_overview', 'host_name', 'host_since',
                     'host_location', 'host_about', 'host_response_time',
                      'host_response_rate', 'host_acceptance_rate', 'host_is_superhost',
                       'host_listings_count', 'host_total_listings_count',
                        'host_verifications', 'host_identity_verified', 'neighbourhood_cleansed',
                         'neighbourhood_group_cleansed', 'latitude', 'longitude', 'property_type',
                          'room_type', 'accommodates', 'bathrooms', 'bathrooms_text', 'bedrooms',
                           'beds', 'amenities', 'price', 'minimum_nights', 'maximum_nights',
                            "minimum_nights_avg_ntm", 'maximum_nights_avg_ntm', 'availability_30',
                             'availability_60', 'availability_90', 'availability_365', 'number_of_reviews',
                              'number_of_reviews_l30d', 'first_review', 'license', 'review_scores_rating',
                               'review_scores_location', 'review_scores_cleanliness', 'review_scores_communication',
                                'review_scores_value', 'instant_bookable', 'reviews_per_month' ]

data2 = data.loc[:,columns]

data2
```

```python
# ahora utilizo la funcion isnull() para saber si hay valores nulos en mi dataframe
valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
# Se reemplazan los valores nulos por valores por defecto en todas las columnas con valores nulos de tipo string
# dando asi un valor por defecto a los valores nulos para identificarlos facilmente
data2["description"] = data2["description"].fillna("Descripción no proporcionada")
data2["neighborhood_overview"] = data2["neighborhood_overview"].fillna("Información acerca del vecindario no proporcionada")
data2["host_name"] = data2["host_name"].fillna("Sin nombre")
data2["host_since"] = data2["host_since"].fillna("20/06/2017")
data2["host_location"] = data2["host_since"].fillna("Información acerca del anfitrión no proporcionada")
data2["host_about"] = data2["host_about"].fillna("Información acerca del anfitrión del no proporcionada")
data2["host_verifications"] = data2["host_verifications"].fillna("['email', 'phone']")
data2["license"] = data2["license"].fillna("Sin licencia")

valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
# Se reemplazan los valores nulos por valores por defecto en todas las columnas con valores nulos de tipo string
# dando asi un valor por defecto a los valores nulos para identificarlos facilmente
data2["host_response_time"] = data2["host_response_time"].fillna("unknown")
data2["host_response_rate"] = data2["host_response_rate"].fillna("0%")
data2["host_acceptance_rate"] = data2["host_acceptance_rate"].fillna("0%")
data2["host_is_superhost"] = data2["host_is_superhost"].fillna("f") # puede ser unknown
data2["host_identity_verified"] = data2["host_identity_verified"].fillna("f") # puede ser unknown
data2["price"] = data2["price"].fillna("$0.00")
data2["first_review"] = data2["first_review"].fillna("01/01/2017")

valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
# Rellenar valores nulos de tipo numerico a traves de metodos estadiscos
# en este caso se rellenan los valores nulos con la media de los valores de la columna y no se utiliza ningun valor por defecto ni bfill o ffill
# para evitar resultados lineales en los datos
data2["host_listings_count"] = data2["host_listings_count"].fillna(round(data2["host_listings_count"].mean(),1))
data2["host_total_listings_count"] = data2["host_total_listings_count"].fillna(round(data2["host_total_listings_count"].mean(),1))
data2["bathrooms"] = data2["bathrooms"].fillna(round(data2["bathrooms"].mean(),1))
data2["bathrooms_text"] = data2["bathrooms_text"].fillna("0 baths")
data2["bedrooms"] = data2["bedrooms"].fillna(round(data2["bedrooms"].mean(),1))
data2["beds"] = data2["beds"].fillna(round(data2["beds"].mean(),1))
data2["review_scores_rating"] = data2["review_scores_rating"].fillna(round(data2["review_scores_rating"].mean(),1))
data2["review_scores_location"] = data2["review_scores_location"].fillna(round(data2["review_scores_location"].mean(),1))
data2["review_scores_cleanliness"] = data2["review_scores_cleanliness"].fillna(round(data2["review_scores_cleanliness"].mean(),1))
data2["review_scores_communication"] = data2["review_scores_communication"].fillna(round(data2["review_scores_communication"].mean(),1))
data2["review_scores_value"] = data2["review_scores_value"].fillna(round(data2["review_scores_value"].mean(),1))
data2["reviews_per_month"] = data2["reviews_per_month"].fillna(round(data2["reviews_per_month"].mean(),1))


valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
# pasamos nuestro dataframe a un archivo .csv
data2.to_csv("madrid_spain_clean.csv")
```
